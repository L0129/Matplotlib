### Bargraphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.style.use('default')
```

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='red')
```

```python
# horizontal bar chart
plt.barh(colors,children,color='purple')
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2017'])
plt.show()
```

```python
# multiple bargraph 
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='black')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='purple')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='pink')

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
# overlapping problem
children = [10,20,40,10,30]
colors = ['something big data and data','blue data problems','green dataset overlappping problems','yelllow dataset overlapping problems','pink dataset overlapping problems']

plt.bar(colors,children,color='purple')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017',color="pink")
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016',color="purple")
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015',color="black")

plt.legend()
plt.show()
```

### Histogram

### Piechart

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
df=pd.read_csv("gayle-175.csv")
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['pink','purple','green','orange','green','red'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True,colors=['pink','purple','green','orange','green','red'])
plt.show()
```
