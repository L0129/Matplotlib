### Scatter Graph

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.style.use('default')
```

```python
df=pd.read_csv("batter.csv")
df=df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'])
```

```python
# label and title
plt.scatter(df['avg'],df['strike_rate'])
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
#color plotting
plt.scatter(df['avg'],df['strike_rate'],color="#FC00D6")
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
plt.scatter(df['avg'],df['strike_rate'],color="#FC00D6",linewidth=7)
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')# linewidth
```

```python
#markers
plt.scatter(df['avg'],df['strike_rate'],color="black",linewidth=1,marker="*")
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
#legend and labels
plt.scatter(df['avg'],df['strike_rate'],color="black",linewidth=1,marker="*",label="top 50 batsman")
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
plt.legend()
```

```python
#grid
plt.scatter(df['avg'],df['strike_rate'],color="black",linewidth=1,marker="*",label="top 50 batsman")
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
plt.legend()
plt.grid()
```
