```python
df=pd.read_csv("sharma-kohli.csv")
df
```

```python
plt.plot(df['index'],df['V Kohli'])
```

```python
### ploting Multiple Plots
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
```

```python
# labels title
plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Colors in plotting
plt.plot(df['index'],df['V Kohli'],color="black")
plt.plot(df['index'],df['RG Sharma'],color="red")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors in hex
plt.plot(df['index'],df['V Kohli'],color="#D9F10F")
plt.plot(df['index'],df['RG Sharma'],color="#FC00D6")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#linestyling 
plt.plot(df['index'],df['V Kohli'],color="#D9F10F")
plt.plot(df['index'],df['RG Sharma'],color="#FC00D6",linestyle="dotted")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(df['index'],df['V Kohli'],color="#D9F10F")
plt.plot(df['index'],df['RG Sharma'],color="#FC00D6",linestyle="dashdot")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
#linewidth
plt.plot(df['index'],df['V Kohli'],color="#D9F10F",linestyle="solid",linewidth=4)
plt.plot(df['index'],df['RG Sharma'],color="#FC00D6",linestyle="dashdot",linewidth=3)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# markers(Diamond->D)
plt.plot(df['index'],df['V Kohli'],color="#D9F10F",linestyle="solid",linewidth=4,marker="D")
plt.plot(df['index'],df['RG Sharma'],color="#FC00D6",linestyle="dashdot",linewidth=3,marker="D")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Marker size
plt.plot(df['index'],df['V Kohli'],color="#D9F10F",linestyle="solid",linewidth=4,marker="o",markersize=8)
plt.plot(df['index'],df['RG Sharma'],color="#FC00D6",linestyle="dashdot",linewidth=3,marker="o",markersize=5)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Legend and label
plt.plot(df['index'],df['V Kohli'],color="#D9F10F",linestyle="solid",linewidth=4,marker="o",label="Virat")
plt.plot(df['index'],df['RG Sharma'],color="#FC00D6",linestyle="dashdot",linewidth=3,marker="o",label="Rohit")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```

```python
# location of label
plt.plot(df['index'],df['V Kohli'],color="#D9F10F",linestyle="solid",linewidth=4,marker="o",label="Virat")
plt.plot(df['index'],df['RG Sharma'],color="#FC00D6",linestyle="dashdot",linewidth=3,marker="o",label="Rohit")

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc="upper center")
```
